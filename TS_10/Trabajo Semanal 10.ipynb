## Inicializacion de bibliotecas

```python
import sympy as sp
import numpy as np

from pytc2.sintesis_dipolo import foster, cauer_LC
from pytc2.dibujar import dibujar_foster_serie, dibujar_foster_derivacion, dibujar_cauer_LC
from pytc2.general import print_latex, print_subtitle, a_equal_b_latex_s
```

# Ejercicio 1
Sea la función:

\begin{equation} Z(s) = \frac{(s^2+3)(s^2+1)}{s(s^2+2)} \end{equation}

Se pide hallar la topología circuital y los valores de los componentes para:

a) Síntesis de Z(s) mediante el método de Foster en su versión  "paralelo" o "derivación".

b) Idem a) mediante Cauer 1 y 2.

## a) Foster en Derivación

### Resolucion en papel

<img src="./img/Ej1_foster1.png" width="650" /> 

### Resolucion simbolica

```python
# Resolución simbólica
s = sp.symbols('s ', complex=True)

# --------------- Función de excitación ---------------
FF = ( (s**2 + 3)*(s**2 + 1) )/( (s**2 + 2) * s )
FF_y = 1/FF

print_latex(a_equal_b_latex_s('Z(s)', FF))
print_latex(a_equal_b_latex_s('Y(s)', FF_y))


# --------------- Foster Derivación ---------------
print_subtitle('Foster Derivación')
k0, koo, ki_wi, _, FFy_foster = foster(FF_y)
print_latex(a_equal_b_latex_s(a_equal_b_latex_s('Y(s)', FF_y)[1:-1], FFy_foster ))
dibujar_foster_derivacion(k0, koo, ki_wi, y_exc = FF_y)
```

## a) Foster Serie

### Resolucion en papel

<img src="./img/Ej1_foster2.png" width="650" /> 

### Resolucion simbolica

```python
# --------------- Foster Serie ---------------
print_subtitle('Foster Serie')
k0, koo, ki_wi, _, FF_foster = foster(FF)
print_latex(a_equal_b_latex_s(a_equal_b_latex_s('Z(s)', FF)[1:-1], FF_foster ))
dibujar_foster_serie(k0, koo, ki_wi, z_exc = FF)
```

## b) Cauer 1: Remociones en infinito

### Resolucion en papel

<img src="./img/Ej1_cauer1.png" width="650" /> 

### Resolucion simbolica

```python
# --------------- Cauer 1: Remociones en infinito ---------------
print_subtitle('Cauer 1: Remociones en infinito')
koo, F_cauer_oo, rem = cauer_LC(FF, remover_en_inf=True)
print_latex(a_equal_b_latex_s(a_equal_b_latex_s('F(s)', FF)[1:-1], F_cauer_oo ))
dibujar_cauer_LC(koo, z_exc = F_cauer_oo)
```

## b) Cauer 2: Remociones en cero

### Resolucion en papel

<img src="./img/Ej1_cauer2.png" width="650" /> 

### Resolucion simbolica

```python
# --------------- Cauer 2: Remociones en cero ---------------
print_subtitle('Cauer 2: Remociones en cero')
k0, F_cauer_0, rem = cauer_LC(FF, remover_en_inf=False)
print_latex(a_equal_b_latex_s(a_equal_b_latex_s('F(s)', FF)[1:-1], F_cauer_0 ))
dibujar_cauer_LC(k0, z_exc = F_cauer_0)
```
