Se observa que la transferencia normalizada correspondiente para el filtro solicitado es:

\begin{equation} T(s) = \frac{ s \frac{1}{2,654} }{ s^2 + s \frac{1}{2,654} + 1^2 } \end{equation}

Desnormalizando para la frecuencia central solicitada de 6kHz:

\begin{equation} T(s) = \frac{ s \frac{6000\cdot2\cdot\pi}{2,654} }{ s^2 + s \frac{6000\cdot2\cdot\pi}{2,654} + (6000\cdot2\cdot\pi)^2 } \end{equation}

\begin{equation} T(s) = \frac{ s \frac{37700}{2,654} }{ s^2 + s \frac{37700}{2,654} + 37700^2 } \end{equation}

## Simulacion numerica

```python
import sympy as sp
import numpy as np
import scipy.signal as sig
from pytc2.sistemas_lineales import analyze_sys

# w ya desnormalizada
w = 1*2*3.1416*6000
Q = 2.654

num= [w/Q, 0]
den= [1, w/Q, w*w]

tf_bp = sig.TransferFunction(num, den)
analyze_sys( tf_bp, 'BPF' )
```

Se observa que el filtro diseñado cumple con lo pedido.

## Diseño con UAF42

A partir de la transferencia obtenida, se busco realizar la implementacion circuital utilizando el circuito integrado UAF42. El UAF42 es un filtro universal cuya topologia permite implementar transferencias pasabajos, pasaaltos o pasabanda de segundo orden, y permite versatilidad a la hora de ajustar los parametros del filtro.

En el anexo se puede encontrar la datasheet completa del integrado, y a continuacion se muestran el circuito y las ecuaciones de diseño provistas por el fabricante:

<img src="./img/UAF1.png"  width=650 /> 
<img src="./img/UAF2.png"  width=650 /> 

Para esta implementacion, las ecuaciones que resultan de interes son la (1), (2) y (6) que corresponden a la frecuencia central, el Q y la ganancia en la banda de paso del pasabanda respectivamente.

#### Frecuencia central

De la ecuacion (1), se observa que:

\begin{equation} w_o^2 = \frac{R_2}{R_1 R_{F1} R_{F2} C_1 C_2} \end{equation}

Considerando que $R_1 = R_2 = 50k\Omega$, $C_1 = C_2 = 1nF$ y adoptando $R_{F1} = R_{F2}$ resulta:

\begin{equation} w_o = \frac{1}{R_F C} \end{equation}

Siendo que se busca $w_o = 2\pi6kHz$ resultan:

\begin{equation} R_{F1} = R_{F2} = 26525\Omega \end{equation}

#### Ganancia en banda de paso

De la ecuacion (6), se observa que:

\begin{equation} A_{BP} = \frac{R_4}{R_G} \end{equation}

Se busca una ganancia de 0dB en la banda de paso, por ende resulta $R_G = R_4 = 50k\Omega$

#### Parametro Q

De la ecuacion (2), se observa que:

\begin{equation} Q = \frac{1 + \frac{R_4 (R_G + R_Q) }{R_G R_Q} }{ 1 + \frac{R_2}{R_1}} \sqrt{ \frac{R_2 R_{F1} C_1}{R_1 R_{F2} C_2} } \end{equation}

Si se consideran los valores ya encontrados para las resistencias, la ecuacion (2) se reduce a:

\begin{equation} Q = \frac{1+\frac{(R_G + R_Q)}{R_Q}}{2} \end{equation}

De la cual finalmente se obtiene que $R_Q = 15424\Omega$

### Valores de las resistencias normalizadas

Finalmente, utilizando resistencias de valores comerciales con tolerancia del 5%, resultan:

* $R_{F1} = R_{F2} = 27k\Omega$
* $R_G = 47k\Omega$
* $R_Q = 15k\Omega$

Al haber alterado los valores de las resistencias, se recalcularon los parametros del filtro:

* $f_o = 5,89 kHz$
* $Q = 2,945 $
* $A_{BP} = 0,592dB$ 

Las diferencias con los valores especificados no resultan sustanciales para el adecuado funcionamiento del filtro, por ende se procedió con el diseño.

## Resultados del analizador de espectro

A continuacion se observan los resultados obtenidos conectando el circuito al analizador de espectro, y realizando un barrido en frecuencia de 100 puntos desde 600 Hz hasta 60 kHz (las tablas con los valores se encuentran en el anexo a fin de facilitar la lectura del documento):

```python
modulo = pd.read_csv("./Analizador/Sweep Data_0_modulo.csv")
fase   = pd.read_csv("./Analizador/Sweep Data_1_fase.csv")

modulo.plot(kind='line', x='X (Hz)',y='Ch-1 (dBr)', logx=True, grid=True, title="Módulo")
fase.plot(kind='line', x='X (Hz)',y='Ch-1 (deg)', logx=True, grid=True, title="Fase analizador")

# El analizador entrega la fase cortada debido a limitaciones tecnicas de su capacidad de graficar, por ende para visualizarla
# correctamente hace falta hacer una correccion:

fase_corregida = fase
fase_corregida.loc[50:99, ['Ch-1 (deg)']] = fase.loc[50:99, ['Ch-1 (deg)']] - 360
fase_corregida.plot(kind='line', x='X (Hz)',y='Ch-1 (deg)', logx=True, grid=True, title="Fase corregida")
```

Es prudente señalar dos diferencias notables entre los graficos anteriores con los graficos realizados a partir de las mediciones con osciloscopio.

* En primer lugar, se observa que en el grafico de modulo la respuesta se haya graficada en dBr en vez de dB. Esto implica que esta tomando el valor en dB relativo a un valor que toma como referencia, en este caso la respuesta en la frecuencia de paso de 6kHz, al que le asigna 0 dB. Luego, todos los valores deberan presentar el offset determinado por el valor real en dB a 6kHz. Este valor corresponde a 0.72 dB segun se observa en la simulacion circuital y en las mediciones con osciloscopio. 

     * Para verificar esto, se toma por ejemplo el valor de la respuesta de modulo a 4kHz entregado por el analizador de espectro y el obtenido con el osciloscopio: -7.74 dB para el analizador y -7.24 mediante el osciloscopio. Se observa una diferencia suficientemente similar a los 0.72 dB.  

* En segundo lugar, se observa que el analizador entrega la fase cortada debido a limitaciones tecnicas de su capacidad de graficar. El rango de la escala va de 180º a -180º, por ende resulta imposible visualizar el resultado esperado: una transicion de -90º a -270º. El analizador presenta un grafico que comienza en -90º, y cuando el valor de fase es menor a -180º salta a 180º. Para corregir esta limitacion, en el codigo anterior se tomaron los valores a partir de los cuales realiza ese salto, y se les restaron 360º. El resultado fue el grafico corregido presentado.
