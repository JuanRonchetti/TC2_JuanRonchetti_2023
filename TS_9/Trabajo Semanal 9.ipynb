# Parte 1 
## Diseño de Filtro IIR

```python
fs = 1000
nyq = fs/2

#------ Plantilla ------

wp = np.array([3, 25]) / nyq
ws = np.array([1, 35]) / nyq
gpass = 0.01
gstop = 40

#------ Disenio ------

num, den = sig.iirdesign(wp, ws, gpass, gstop, analog=False, ftype='butter', output='ba')
sos_iir = sig.iirdesign(wp, ws, gpass, gstop, analog=False, ftype='butter', output='SOS')

wrad, hh = sig.sosfreqz(sos_iir, worN=1000)
ww = wrad/np.pi * nyq

Wgd, gd = sig.group_delay((num, den))
wwgd = Wgd/np.pi * nyq

t = np.arange(0, 10, 0.01)
imp = np.fft.ifft(hh)

#------------------------------- Graficos -------------------------------

#tf = sig.TransferFunction(num, den, dt=1/fs)
#analyze_sys(tf)

#------ Modulo ------
modulo = plt.figure()
plt.title('Filtro IIR: Modulo')
plt.plot(ww, 20 * np.log10(abs(hh)))
plt.xlabel('Frecuencia [Hz]')
plt.ylabel('Modulo [dB]')
plt.grid()
plt.axis([0,100,-50, 5])
plt.show()

#------ Fase ------
fase = plt.figure()
plt.title('Filtro IIR: Fase')
angles = np.unwrap(np.angle(hh))
plt.plot(ww, angles)
plt.ylabel('Fase [rads]')
plt.xlabel('Frecuencia [Hz]')
plt.grid()
plt.axis([0,100,-50, 5])
plt.show()

#------ Retardo ------
ret = plt.figure()
plt.title('Filtro IIR: Retardo')
plt.plot(wwgd, gd)
plt.ylabel('Retardo [s]')
plt.xlabel('Frecuencia [Hz]')
plt.grid()
plt.show()

#------ Respuesta al impulso ------
impulso = plt.figure()
plt.title('Filtro IIR: Respuesta al Impulso')
plt.plot(t, imp)
plt.ylabel('Amplitud')
plt.xlabel('Tiempo [s]')
plt.grid()
plt.show()
```

# Parte 2
## Levantamiento de datos de ECG

```python
mat_struct = sio.loadmat('ecg.mat')
ecg_one_lead = mat_struct['ecg_lead']
ecg_one_lead = ecg_one_lead.flatten()
cant_muestras = len(ecg_one_lead)

fig_sz_x = 10
fig_sz_y = 7
fig_dpi = 100 # dpi
```

## Filtrado con filtro IIR

```python
ECG_f_iir = sig.sosfilt(sos_iir, ecg_one_lead)

demora_iir = 115

# Segmentos de interés con ALTA contaminación

regs_interes = ( 
        np.array([5, 5.2]) *60*fs, # minutos a muestras
        np.array([12, 12.4]) *60*fs, # minutos a muestras
        np.array([15, 15.2]) *60*fs, # minutos a muestras
        )

for ii in regs_interes:
    
    # intervalo limitado de 0 a cant_muestras
    zoom_region = np.arange(np.max([0, ii[0]]), np.min([cant_muestras, ii[1]]), dtype='uint')
    
    plt.figure(figsize=(fig_sz_x, fig_sz_y), dpi= fig_dpi, facecolor='w', edgecolor='k')
    plt.plot(zoom_region, ecg_one_lead[zoom_region], label='ECG', linewidth=2)
    plt.plot(zoom_region, ECG_f_iir[zoom_region + demora_iir], label='Filtrado IIR')
    #plt.plot(zoom_region, ECG_f_win[zoom_region + demora], label='Win')
    
    plt.title('ECG filtering example from ' + str(ii[0]) + ' to ' + str(ii[1]) )
    plt.ylabel('Adimensional')
    plt.xlabel('Muestras (#)')
    
    axes_hdl = plt.gca()
    axes_hdl.legend()
    axes_hdl.set_yticks(())
            
    plt.show()
```
