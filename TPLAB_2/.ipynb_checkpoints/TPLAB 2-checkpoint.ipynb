### FIR Least Squares

Para el diseño del FIR mediante la aproximación de cuadrados mínimos se utilizaron las funciones de la biblioteca signal de scipy. 

A continuacion se detalla el codigo mediante el cual se obtuvo el filtro, y el resultado obtenido:

```python
fs = 40000
nyq = 20000

cant_coef = 77

fpass = np.array([2000, 8000])
fstop = np.array([4000, 6000])
gpass = 1
gstop = 20 

frecs = [0.0,   fpass[0],     fstop[0],     fstop[1],     fpass[1],     nyq]
gains = [0,     0,     -gstop,     -gstop,      0,      0] 
gains = 10**(np.array(gains)/20)

fir_ls = sig.firls(cant_coef, frecs, gains, fs=fs)

np.set_printoptions(precision=20, suppress = True)
print("Coeficientes del filtro: \n")
print(fir_ls)

wrad, hh = sig.freqz(fir_ls, worN = 1000)
ww = wrad/np.pi * nyq

#------ Modulo ------
modulo = plt.figure()
plt.title('FIR LS: Modulo')
plt.plot(ww, 20 * np.log10(abs(hh)))
plt.xlabel('Frecuencia [Hz]')
plt.ylabel('Modulo [dB]')
plt.grid()
plt.show()

#------ Fase ------
fase = plt.figure()
plt.title('FIR LS: Fase')
angles = np.unwrap(np.angle(hh))
plt.plot(ww, angles)
plt.ylabel('Fase [rads]')
plt.xlabel('Frecuencia [Hz]')
plt.grid()
plt.show()
```

## Filtro antialias analógico

Para poder ingresar la señal proveniente del generador de funciones correctamente al MCU, se diseñó un filtro pasabajos antialias analogico.

### Diseño

Se decidió implementar un filtro pasabajos con la topología Sallen-Key, con una $f_c = \frac{40kHz}{2} = 20kHz$.

A partir de las siguientes ecuaciones de diseño para el Sallen-Key se llegó al valor de los componentes necesarios:

\begin{equation} T(s) = \frac{ \frac{1}{R_1 R_2 C_1 C_2} }{ s^2 + s (\frac{1}{R_1 C_1} + \frac{1}{R_2 C_1}) + \frac{1}{R_1 R_2 C_1 C_2} } \end{equation}

\begin{equation} w_o^2 = \frac{1}{R_1 R_2 C_1 C_2} \end{equation}

Tomando:

* $R_1 = R_2 = 6800 \Omega$
* $C_1 = C_2 = 1n2 F$

Se obtiene la $f_c$ buscada:

\begin{equation} f_c = \frac{1}{2 \pi \sqrt{R_1 R_2 C_1 C_2}} = 19504 Hz \end{equation}

Resultando asi en la siguiente transferencia:

\begin{equation} T(s) = \frac{ 15018262206 }{ s^2 + s 245098 + 15018262206 } \end{equation}

# Análisis de los resultados

Se observa que en las dos experiencias de laboratorio, utilizando dos metodos de medicion distintos, se llegó a las mismas transferencias obtenidas al diseñar los filtros.

Existe una diferencia apreciable en cuanto a la calidad de los graficos obtenidos mediante el analisis manual con osciloscopio, y los obtenidos mediante el barrido con el analizador de audio. Esto es atriuible principalmente a 2 factores: por un lado a un error humano a la hora operar el osciloscopio y seleccionar las frecuencias precisamente, y por otro lado la diferencia en la cantidad de puntos tomados. 

Se observa que la respuesta de modulo correspondiente al Talkthrough, es decir, la salida sin que el MCU realice ningun filtrado, corresponde a una respuesta del tipo pasabajos. Esto es debido a los filtros analogicos aplicados a la entrada y a la salida del MCU.

<img src="./img/Analog_resp.png"  width=450 /> 

Como se observa en la imagen, el filtro analogico diseñado posee una atenuacion a los 10kHz. Si se considera que se tienen dos filtros identicos en cascada, y dada la dispersion de los componenentes, la atenuacion se vuelve mas severa, resultando la causante del efecto pasabajos observado en el talkthrough.

Este efecto es arrastrado hacia las demas respuestas de los filtros, siendo claramente observable en el FIR de cuadrados minimos elimina-banda, donde se observa que luego de la banda de eliminacion la respuesta decrece.
